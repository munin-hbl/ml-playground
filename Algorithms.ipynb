```python
#fizz buzz algo

def fizzbuzz(n):
    for n in range(n):
        print("Fizz"*(not n % 3) + "Buzz"*(not n % 5) or n)
fizzbuzz(20)
```

```python
#Tree array
class Node:
    def __init__(self, data):
        self.data = data
        self.child = None
        self.next = None


def print_tree_r(node, depth=0):
    while node:
        # Hvis noden har et barn, print og gå rekursivt ned
        if node.child:
            print(" " * (depth * 3) + "{")
            print_tree_r(node.child, depth + 1)
            print(" " * (depth * 3) + "}")

        # Print selve nodens data
        print(" " * (depth * 3) + str(node.data))

        # Gå videre til næste søskende
        node = node.next


def print_tree(root):
    print_tree_r(root, 5)
```
